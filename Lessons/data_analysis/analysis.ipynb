## `read_csv_rows`

- Function name: `read_csv_rows`
- Parameters: `str` path to csv file
- Return type: `list[dict[str,str]]`

This function reads csv files and gives me a data structure to work wiht.

```python
def read_csv_rows(filename: str) -> list[dict[str,str]]:
    """Read the rows of a CSV."""
    result: list[dict[str,str]] = []
    
    file_handle = open(filename, "r", encoding = "utf8")
    # open() is a built-in function that opens files
    #"r" says that we are just opening csv to read data.
    # encoding tells python how to interpret the text.

    #Make a "DictReader"
    csv_reader = DictReader(file_handle)

    for row in csv_reader:
        result.append(row)
        # now each row makes the unique dictionary that we have just defined.

    # close the file
    file_handle.close()

    return result
```
